### El modelo Lotka-Volterra es un conjunto de ecuaciones matemáticas que describe la interacción dinámica entre dos especies: una especie presa y una especie depredadora. Es un modelo matemático que utiliza ecuaciones diferenciales para representar cómo cambian las poblaciones de presas y depredadores a lo largo del tiempo debido a sus interacciones.

### ¿Para qué sirve?
- Estudiar Ecosistemas: Ayuda a entender cómo las poblaciones de presas y depredadores fluctúan y se afectan mutuamente.
- Predecir Dinámicas Poblacionales**: Permite predecir ciclos periódicos en las poblaciones y analizar la estabilidad de los ecosistemas.
- Gestión y Conservación: Proporciona una base para diseñar estrategias de manejo y conservación de especies en la naturaleza.

## Modelo No.1

```python
import matplotlib.pyplot as plt
from random import *
from numpy import *
```

```python
# Parámetros del modelo
a = 1.0  # Aumenta la tasa de crecimiento de las presas
b = 0.5
c = 0.3
e = 0.2
dt = 0.001; max_time = 100

# Tiempo inicial y poblaciones
t = 0
x = 1.0  # Población de presas
y = 0.5  # Población de depredadores

# Listas vacías para almacenar tiempo y poblaciones
t_list = []
x_list = []
y_list = []

# Inicializar listas
t_list.append(t)
x_list.append(x)
y_list.append(y)

while t < max_time:
    # Calcular nuevos valores para t, x, y
    t = t + dt
    x = x + (a * x - b * x * y) * dt
    y = y + (-c * y + e * x * y) * dt

    # Almacenar nuevos valores en listas
    t_list.append(t)
    x_list.append(x)
    y_list.append(y)

# Graficar los resultados    
plt.plot(t_list, x_list, 'r', label="Población de presas", linewidth=2)
plt.plot(t_list, y_list, 'g', label="Población de depredadores", linewidth=2)

# Añadir etiquetas y título
plt.xlabel("Tiempo")
plt.ylabel("Población")
plt.title("Modelo Presa-Depredador")

# Añadir una leyenda
plt.legend()

# Mostrar la gráfica
plt.show()
```

Comportamiento: La población de presas (línea roja) crece rápidamente debido al mayor valor de `a`, lo que provoca una mayor población de depredadores (línea verde) después de un pequeño desfase. La alta tasa de crecimiento inicial de las presas genera oscilaciones de mayor amplitud, donde ambas poblaciones alcanzan valores más altos y luego disminuyen más drásticamente.
